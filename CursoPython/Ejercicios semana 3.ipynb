```python
#12_Linear_regression
!Pip install pandas
!Pip install scipy
!Pip install seaborn
!Pip install matplotlib

import pandas as pd
from scipy.stats import linregress
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos
path = "C:/Users/medra/OneDrive/Documentos/Fernanda/Cursos/GitHub/Curso_Python_Fer/Modelab-Python/data/Terminos_lagoon_TA_DIC__2023_RawData.csv"
CO2Data = pd.read_csv(path)

# SelecciÃ³n de variables
Temp = CO2Data['Temp_C'].dropna()
Salinity = CO2Data['Sal_psu'].dropna()
DIC = CO2Data['DIC_micromol_kg'].dropna()

# 1ï¸â£ Salinity vs Temperature
slope1, intercept1, r_value1, p_value1, std_err1 = linregress(Temp, Salinity)
print("Linear Regression: Salinity vs Temperature")
print(f"Slope: {slope1}, Intercept: {intercept1}, RÂ²: {r_value1**2}, p-value: {p_value1}")

# GrÃ¡fico
plt.figure(figsize=(6,5))
sns.regplot(x=Temp, y=Salinity)
plt.title("Salinity vs Temperature")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Salinity (PSU)")
plt.show()

# 2ï¸â£ DIC vs Salinity
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(Salinity, DIC)
print("\nLinear Regression: DIC vs Salinity")
print(f"Slope: {slope2}, Intercept: {intercept2}, RÂ²: {r_value2**2}, p-value: {p_value2}")

# GrÃ¡fico
plt.figure(figsize=(6,5))
sns.regplot(x=Salinity, y=DIC)
plt.title("DIC vs Salinity")
plt.xlabel("Salinity (PSU)")
plt.ylabel("DIC (Âµmol/kg)")
plt.show()
```

```python
#13_spearman_correlation
import pandas as pd
from scipy.stats import pearsonr

# FunciÃ³n para calcular Pearson correlation
def pearson_correlation(x, y):
    """
    Calcula la correlaciÃ³n de Pearson entre dos variables.
    
    ParÃ¡metros:
    x, y : listas, arrays o Series de pandas
    
    Retorna:
    r : coeficiente de correlaciÃ³n de Pearson
    p : p-value
    """
    r, p = pearsonr(x, y)
    return r, p

# Datos
path = "C:/Users/medra/OneDrive/Documentos/Fernanda/Cursos/GitHub/Curso_Python_Fer/Modelab-Python/data/Terminos_lagoon_TA_DIC__2023_RawData.csv"
CO2Data = pd.read_csv(path)

# SelecciÃ³n de variables
Temp = CO2Data['Temp_C'].dropna()
Salinity = CO2Data['Sal_psu'].dropna()
DIC = CO2Data['DIC_micromol_kg'].dropna()

# Pearson correlation
r_temp_sal, p_temp_sal = pearson_correlation(Temp, Salinity)
r_dic_sal, p_dic_sal = pearson_correlation(DIC, Salinity)

print(f"Pearson correlation Temperature vs Salinity: r={r_temp_sal:.3f}, p={p_temp_sal:.3e}")
print(f"Pearson correlation DIC vs Salinity: r={r_dic_sal:.3f}, p={p_dic_sal:.3e}")
```
